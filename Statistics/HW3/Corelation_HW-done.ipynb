```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from statsmodels.sandbox.regression.predstd import wls_prediction_std
```

```python
data = pd.read_csv('water.csv')
data.head()
```

```python
plt.plot(data['mortality'], label = 'deaths')
plt.plot(data['hardness'], label = 'water hardness')
plt.legend()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, label = "1", frame_on = False)
ax2 = fig.add_subplot(111, label = "2", frame_on = False)
ax.plot(data['mortality'], color = 'b')
ax2.plot(data['hardness'], color = 'r')
ax2.yaxis.tick_right()
ax.legend(['deaths'])
ax2.legend(['water hardness'])
```

```python
plt.scatter(data['mortality'], data['hardness'])
```

```python
data[['mortality', 'hardness']].corr(method = 'pearson')
```

```python
data[['mortality', 'hardness']].corr(method = 'spearman')
```

```python
X = data[['mortality']].values
Y = data['hardness'].values
model = LinearRegression()
model.fit(X, Y)
print(model.score(X, Y))
print(model.coef_)
print(model.intercept_)
```

```python
plt.scatter(data['mortality'], data['hardness'])
plt.plot(data['mortality'], model.predict(data[['mortality']]))
```

```python
X = sm.add_constant(X)
model = sm.OLS(Y, X)
results = model.fit()
print(results.summary())
```

```python
print('Коэффициент детерминации R2:', results.rsquared)
```

```python
plt.scatter(X[:, 1], results.resid)
```

```python
plt.hist(results.resid, color = 'r')
```
