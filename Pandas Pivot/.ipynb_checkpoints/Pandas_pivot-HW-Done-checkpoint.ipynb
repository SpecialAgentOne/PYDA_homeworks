## Задание 1

Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:

    если источник traffic_source равен yandex или google, то в source_type ставится organic
    для источников paid и email из России - ставим ad
    для источников paid и email не из России - ставим other
    все остальные варианты берем из traffic_source без изменений



```python
# "Компактный вариант" - спасибо google!

import pandas as pd

data = pd.read_csv('visit_log.csv', sep=';')
#data.head()

data['source_type'] = data['traffic_source']
data.loc[data['traffic_source'].isin(['yandex','google']), 'source_type'] = 'organic'
data.loc[(data['region'] == 'Russia') & (data['traffic_source'].isin(['paid','email'])),'source_type'] = 'ad'
data.loc[(data['region'] != 'Russia') & (data['traffic_source'].isin(['paid','email'])),'source_type'] = 'other'

data.head()
```

```python
#Вариант решения "долгий", на основе лекции

import pandas as pd

data = pd.read_csv('visit_log.csv', sep=';')
#data.head()

data[ data['visit_id'].str.contains('3c', case = False) ].head()

src1 = ['yandex', 'google']
src2 = ['paid', 'email']

for traffic_source in src1: 
    data.loc[data.traffic_source == traffic_source, 'traffic_type'] = 'organic' 
for traffic_source in src2: 
    data.loc[(data.traffic_source == traffic_source) & (data.region == 'Russia'), 'traffic_type'] = 'ad' 
for traffic_source in src2: 
    data.loc[(data.traffic_source == traffic_source) & (data.region != 'Russia'), 'traffic_type'] = 'other' 
data.loc[pd.isnull(data.traffic_type), 'traffic_type'] = data.traffic_source

result = data[data['visit_id'].str.contains('3c', case = False)]  
result.head()
```

## Задание 2
В файле URLs.txt содержатся url страниц новостного сайта. Вам необходимо отфильтровать его по адресам страниц с текстами новостей. Известно, что шаблон страницы новостей имеет внутри url следующую конструкцию: /, затем 8 цифр, затем дефис. Выполните следующие действия:

Прочитайте содержимое файла с датафрейм
Отфильтруйте страницы с текстом новостей, используя метод str.contains и регулярное выражение в соответствии с заданным шаблоном

```python
import pandas as pd

url = pd.read_csv('URLs.txt')

regex = '/[0-9]{8}-'

result = url[url['url'].str.contains(regex, regex = True)]
result.head()
```

## Задание 3

В датафрейме data создайте столбец lemmas, в котором вычислите леммы поисковых запросов из столбца keyword. Леммы должны иметь строковый тип.

```python
import pandas as pd
from pymystem3 import Mystem

data = pd.DataFrame({
    'keyword': ['курс гривны к рублю', 'доллары в рубли', '100 долларов в рублях', 'курс рубля'],
    'shows': [125076, 114173, 97534, 53546],
})

def lemmas_tool(value):
    lemmas = Mystem().lemmatize(value)
    return ' '.join(lemmas).strip()

data['lemmas'] = data['keyword'].apply(lemmas_tool)

data.head()
```
