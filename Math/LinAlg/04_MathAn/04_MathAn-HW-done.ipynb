на отрезке [-5; 5]

```python
x = Symbol('x')
y = ((sin(2*x+1))**5)
first = y.diff(x)
second = first.diff(x)
```

```python
def one(x):
    return 10*sin(2*x + 1)**4*cos(2*x + 1)
def two(x):
    return -20*sin(2*x + 1)**5 + 80*sin(2*x + 1)**3*cos(2*x + 1)**2
```

```python
x = np.linspace(-5, 5, 100)
y = [one(x[i]) for i in range(len(x))]
z = [two(x[i]) for i in range(len(x))]  
```

```python
plt.plot(x, y)
plt.plot(x, z)

plt.title('Graph')

plt.show()
```
